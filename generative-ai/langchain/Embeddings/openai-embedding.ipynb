#### Workflow Example from Data Ingestion to Embeddings

1. Data Ingestion (Loading Text from Data Source)
2. Data Transformer (Splitting Text into Text Chunks)
3. Embeddings (Text Chunks to Vectors)


```python
# Data Ingestion
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
# Data Transformer
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
# Embeddings
embeddings = OpenAIEmbeddings(model='text-embedding-3-large')
vectors = embeddings.embed_documents([doc.page_content for doc in final_docs])
vectors
```

```python
# (OR) Combining Embedding technique with Vector StoreDB (Chroma DB)
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_docs, embeddings)
```

```python
# Retrieve results from querying vector storeDB
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
